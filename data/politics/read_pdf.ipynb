```python
dfs = tabula.read_pdf('glos_86_6.pdf', stream = True, multiple_tables=True, pages='all')
len(dfs)
```

```python
df1 = pd.DataFrame(columns=[0,1])
df_add = pd.DataFrame(columns=[0,1])

for i, col in enumerate( dfs[0].columns ):
    
    if i in (0, 1):
        df1[i] = dfs[0][i]
    elif i % 2 == 0:
        target_col = i % 2
        df_add[[0,1]] = dfs[0][[col, col+1]]
        df1 = df1.append( df_add, ignore_index=True )
#        df1[target_col] = pd.concat([df1[target_col], dfs[0][col]], ignore_index=True)
```
